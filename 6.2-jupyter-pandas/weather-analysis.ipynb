# Weather Analysis

```python
import pandas as pd
```

## Load data

```python
df = pd.read_csv("data_small/TG_STAID000001.txt",
                 skiprows=20,
                 parse_dates=["    DATE"],
                 date_format="%Y%m%d", # Only to inform how is the date string format
                 )
```

## Filtering data

```python
# Example on how to use or in conditioning the row filtering: make use of | and ()
# df_filtered = df[(df[" Q_TG"] == 1) | (df[" Q_TG"] == 9)]

df_filtered = df[df[" Q_TG"] == 0]
df_filtered
```

## Replacing data using .loc

```python
df_replaced = df
df_replaced.loc[ ((df_replaced[" Q_TG"] == 1) | (df_replaced[" Q_TG"] == 9)), ("   TG")]= float('nan')
df_replaced
```

## Drop rows with nan

```python
df_replaced.dropna(axis=0)
```

## Basic Statistics

```python
# Adjust the mean by not considering the NaN
print(f'Mean: ', df_replaced.loc[:, "   TG"].mean())
print(f'Mean: ', df_replaced.loc[:, "   TG"].max())
```
